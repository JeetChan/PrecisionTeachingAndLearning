```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///database')
```

```python
with engine.connect() as conn, conn.begin():
    submit_issue_data = pd.read_sql_table('submit_issue', conn)
    stats_code_data = pd.read_sql_table('stats_code', conn)
    issue_info_data = pd.read_sql_table('issue_info', conn)
```

```python
def count_issue(github_user_name):
    num = 0
    for item in issue_info_data['comment_content']:
        s = item.split(",")
        for item in s:
            try:
                if github_user_name in item:
                    num += 1
            except:
                print("user_name: ", github_user_name)
    return num
github_user_name = 'leiyunhe'
count_issue(github_user_name)
```

```python
issue_response = {}
for index in submit_issue_data['github_user_name']:
    issue_response[index] = count_issue(index)
```

```python
len(issue_response)
```

```python
temp = pd.Series(issue_response)
```

```python
temp_pd = pd.DataFrame(temp,columns=['response'])
```
